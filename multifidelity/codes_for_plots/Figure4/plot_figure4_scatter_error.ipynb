```python
import pandas as pd
def to_pd(xs, exp, predicted):
    from collections import defaultdict
    d = defaultdict(list)
    for i, x in enumerate(xs):
        d['x'].append(round(x, 3))
        d['Experiment'].append(round(exp[i], 3))
        for split, pred in enumerate(predicted[i]):
            d['Predicted %d' % split].append(round(pred, 3))
    return pd.DataFrame(d)
```

```python
AlGaN = to_pd(xs, exp, predicted)
```
