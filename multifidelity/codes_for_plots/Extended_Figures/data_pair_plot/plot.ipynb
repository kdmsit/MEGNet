```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.size'] = 22
plt.rcParams['font.family'] = 'Arial'

import pandas as pd

df = pd.read_csv('all_datapairs.csv')

import seaborn as sns
g = sns.pairplot(df,corner=True, diag_kind='kde',
             plot_kws=dict(s=90, color="#19ded3", edgecolor='k', linewidth=0.2, alpha=0.3),
             diag_kws=dict(shade=True, color='#ed0a6c'))

axes = g.axes

for i, ax in enumerate(axes):

    
    for ax_temp in ax:
        if ax_temp:
            ax_temp.set_xlim((0,15))
            ax_temp.set_xticks([0, 5, 10, 15])
            ax_temp.set_ylim((0,15))
        
plt.tight_layout(pad=0.1)
plt.savefig('pair_plots.pdf')
```
