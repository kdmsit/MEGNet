```python
import pickle

with open("BSCF.p", "rb") as f:
    data = pickle.load(f)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['font.size'] = 22
plt.rcParams['font.family'] = "Arial"

avg = [np.mean(i) for i in data['predicted']]
xs = data['x']
deltas = data['delta']

plt.figure(figsize=(6.5, 5))
X, Y = np.meshgrid(xs, deltas)
plt.pcolor(X, Y, np.array(avg).reshape((101, 101)), cmap='hot')
plt.xlabel('$\delta$ in $\mathregular{Ba_{0.5}Sr_{0.5}Co_{x}Fe_{1-x}O_{3-\delta}}$')
plt.ylabel('$x$')
cbar = plt.colorbar()
cbar.set_label('E$_g$ (eV)')
plt.tight_layout()
plt.savefig('BSCF.pdf')
```

```python
import pandas as pd

df = pd.DataFrame(np.array(avg).reshape((101, 101)))

df.columns = ["delta=%.2f" % i for i in deltas]
df.index = ["x=%.2f" % i for i in xs]
```
