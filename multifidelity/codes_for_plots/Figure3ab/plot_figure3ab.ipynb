```python
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
import numpy as np
with open('tsne_final.p', 'rb') as f:
    data = pickle.load(f)
```

```python
f1_tsne = data['f1_tsne']
f2_tsne = data['f2_tsne']
```

```python

plt.figure(figsize=(6.3, 5))
plt.rcParams['font.size'] = 22
plt.rcParams['font.family'] = 'Arial'
arg = np.argsort(data['values'])
plt.scatter(f1_tsne[arg, 0], f1_tsne[arg, 1], c=np.array(data['values'])[arg], 
            s=40, lw=0, cmap='hot', alpha=0.5)
cbar = plt.colorbar()
cbar.set_label('$E_g$ (eV)')
plt.tight_layout()
# plt.savefig('F1_100_0_perplexity_%d.pdf' %p)
```

```python
import pandas as pd

pd.DataFrame({"TSNE1": round(f1_tsne[arg, 0], 3), 
              "TSNE2": round(f1_tsne[arg, 1], 3), "Band gap": np.array(data['values'])[arg]})
```

```python
plt.figure(figsize=(6.3, 5))
plt.rcParams['font.size'] = 22
plt.rcParams['font.family'] = 'Arial'

arg = np.argsort(data['values'])
plt.scatter(f2_tsne[arg, 0], f2_tsne[arg, 1], c=np.array(data['values'])[arg], 
            s=40, lw=0, cmap='hot', alpha=0.5)
cbar = plt.colorbar()
cbar.set_label('$E_g$ (eV)')
plt.tight_layout()
# plt.savefig('F2_100_41000_perplexity_%d.pdf' %p)
```
